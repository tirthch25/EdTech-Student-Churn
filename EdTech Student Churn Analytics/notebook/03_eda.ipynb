```python
#1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
pd.set_option("display.max_columns", None)
```

```python
#2. Load Processed Dataset
df = pd.read_csv("../data/processed/edtech_churn_final.csv")
df.head()
```

```python
#3. Basic Dataset Overview
df.shape
df.info()
df.describe()
```

```python
#4. Overall Churn Rate
churn_rate = df['churn'].mean()
churn_rate
```

```python
#Visualization
sns.countplot(x='churn', data=df)
plt.title("Overall Student Churn Distribution")
plt.show()
```

```python
#5. Engagement vs Churn (MOST IMPORTANT)
#5.1 Total Clicks
sns.boxplot(x='churn', y='total_clicks', data=df)
plt.title("Total Engagement vs Churn")
plt.show()
```

```python
#5.2 Active Days
sns.boxplot(x='churn', y='active_days', data=df)
plt.title("Active Days vs Churn")
plt.show()
```

```python
#5.3 Inactivity Flag
sns.countplot(x='inactive_14_days', hue='churn', data=df)
plt.title("14-Day Inactivity vs Churn")
plt.show()
```

```python
#6. Assessment Performance vs Churn
#6.1 Average Score
sns.boxplot(x='churn', y='avg_score', data=df)
plt.title("Average Assessment Score vs Churn")
plt.show()
```

```python
#6.2 Missed Assessments
sns.boxplot(x='churn', y='missed_assessments', data=df)
plt.title("Missed Assessments vs Churn")
plt.show()
```

```python
#7. Resource Type Engagement vs Churn
#Aggregate resource engagement
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
resource_cols = numeric_cols.drop(['id_student', 'churn'])
resource_mean = df.groupby('churn')[resource_cols].mean().T

resource_mean.plot(kind='bar', figsize=(12,6))
plt.title("Average Numeric Feature Values by Churn")
plt.ylabel("Average Value")
plt.show()
```

```python
#8. Correlation Heatmap
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title("Feature Correlation Heatmap (Numeric Features)")
plt.show()
```
